```python
from pymnet import *
import matplotlib
matplotlib.use('Agg')
```

### Initialize pymnet graph

```python
mnet = MultilayerNetwork(aspects=1)
mnet[1, 'a'][2, 'b'] = 1
mnet[1, 'a'][4, 'a'] = 1
mnet[1, 'a'][5, 'a'] = 1
mnet[1, 'a'][6, 'b'] = 1
mnet[1, 'b'][2, 'b'] = 1
mnet[5, 'b'][4, 'b'] = 1
mnet[4, 'b'][6, 'b'] = 1
mnet[5, 'b'][6, 'b'] = 1
```

```python
draw(mnet,layout="spring")
```

### Multiplex networks

```python
mplex = MultiplexNetwork(couplings=('categorical', 1))
# Add layers
mplex.add_layer('a')
mplex.add_layer('b')
mplex.add_layer('c')
```

```python
mplex[1,'a'][2,'a'] = 1
mplex[4,'a'][5,'a'] = 1
mplex[4,'a'][2,'a'] = 1
mplex[1,'a'][4,'a'] = 1
draw(mplex,layout="spring")
```
